**5. Perform meaningful visualizations using the dataset. Bring at least 3 
visualizations.**

To perform meaningful visualizations using the Iris dataset, using various plotting libraries in Python like pandas, seaborn, and matplotlib. 

```python
# importing necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Bar plot using Pandas:**

```python
import pandas as pd
import matplotlib.pyplot as plt #importing pyplot from matplotlib as plt

iris_df['Classification'].value_counts().plot(kind='bar')
plt.title('Count of Classification')
plt.xlabel('Classification')
plt.ylabel('Count')
plt.show()

#This will create a bar plot showing the count of each Classification in the Iris dataset.
```

**Box plot using seaborn:**

```python
import seaborn as sns #importing seaborn 

sns.boxplot(data=iris_df, x='Classification', y='SL')
plt.title('Distribution of Sepal Length (SL) by Classification')
plt.xlabel('Classification')
plt.ylabel('SL')
plt.show()

#This will create a box plot showing the distribution of sepal length (SL) for each Classification in the Iris dataset.
```

**Histogram using matplotlib:**

```python
import matplotlib.pyplot as plt

plt.hist(iris_df['PW'], bins=10, color='violet')
plt.title('Distribution of Petal Width (PW)')
plt.xlabel('PW')
plt.ylabel('Count')
plt.show()

#This will create a histogram showing the distribution of PW (Petal Width) in the Iris dataset.
```

**Scatter matrix plot using pandas**

```python
# scatter matrix plot for all columns
pd.plotting.scatter_matrix(iris_df, diagonal='kde', alpha=0.8, c='r')
plt.suptitle("Scatter Matrix Plot for Iris Dataset")
plt.show()
```

**Scatter plot using matplotlib**

```python
#scatter plot for Sepal Length (SL) vs Sepal Width (SW)


import matplotlib.pyplot as plt

# Create a dictionary to map the category names to colors
colors = {"Iris-setosa": "red", "Iris-versicolor": "green", "Iris-virginica": "blue"}

# Create the scatter plot with different colors for each category
plt.scatter(iris_df["SL"], iris_df["SW"], c=iris_df["Classification"].apply(lambda x: colors[x]))

# Add labels and title
plt.xlabel("SL")
plt.ylabel("SW")
plt.title("Scatter plot for Sepal Length (SL) vs Sepal Width (SW)")

# Add legend
for category, color in colors.items():
    plt.scatter([], [], c=color, label=category) # Dummy scatter plot to add legend
plt.legend()

plt.show()

 #color_map dictionary maps each flower species to a specific color. 
 #Then the apply function is used to apply this mapping to the "Classification" column of the dataset, so that each flower species is plotted with the corresponding color.
```

**Violin plot using seaborn**

```python
# Create violin plot for Petal Length (PL)

sns.violinplot(x="PL", data=iris_df, color='brown')
plt.title("Violin plot for Petal Length (PL)")
plt.show()
```

**Pairplot using seaborn**

```python
# Create pairplot for all columns

sns.pairplot(data=iris_df)
plt.suptitle("PAIRPLOT FOR IRIS DATASET", y=1.05, fontsize=16)
plt.show()

#  using the suptitle function from matplotlib to add the title at the top of the plot. 
# The y parameter adjusts the vertical position of the title, and the fontsize parameter sets the font size. 
```
