**HANDLING MISSING VALUES**

```python
# Fill missing values for each column based on data type

# Categorical columns
data['gender'].fillna(data['gender'].mode()[0], inplace=True)
data['strong_foot'].fillna(data['strong_foot'].mode()[0], inplace=True)
data['dedication_level'].fillna(data['dedication_level'].mode()[0], inplace=True)
data['coaching'].fillna(data['coaching'].mode()[0], inplace=True)

# Numerical columns
data['ball_controlling_skills'].fillna(data['ball_controlling_skills'].median(), inplace=True)
data['jumping_skills'].fillna(data['jumping_skills'].median(), inplace=True)
data['penalties_conversion_rate'].fillna(data['penalties_conversion_rate'].median(), inplace=True)
data['mental_strength'].fillna(data['mental_strength'].median(), inplace=True)
data['fitness_rating'].fillna(data['fitness_rating'].median(), inplace=True)

# For 'shot_accuracy', a significant number of missing values; dropping this column
data.drop(columns=['shot_accuracy'], inplace=True)

# For 'behaviour_rating', 'matches_played', 'years_of_experience', filling with 0 (or appropriate value)
data['behaviour_rating'].fillna(0, inplace=True)
data['matches_played'].fillna(0, inplace=True)
data['years_of_experience'].fillna(0, inplace=True)

# Dropping rows with missing values in other columns
data.dropna(subset=['age', 'body_reflexes'], inplace=True)

# Confirming no missing values remain
print(data.isnull().sum())
```

**CHECK FOR OUTLIERS**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a box plot for each numerical column
plt.figure(figsize=(16, 10))
sns.boxplot(data=data.drop(['gender', 'country', 'strong_foot', 'dedication_level', 'coaching'], axis=1))
plt.xticks(rotation=90)
plt.title('Box Plots for Numerical Columns')
plt.show()
```

```python
import pandas as pd

# Define a function to count outliers in a column using IQR
def count_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return outliers.shape[0]

# Select only the numerical columns (excluding categorical ones)
numerical_columns = data.select_dtypes(include=['int64', 'float64'])

# Count outliers in each numerical column
outliers_count = numerical_columns.apply(count_outliers)

# Sum the counts of outliers across all columns
total_outliers = outliers_count.sum()

print("Total number of outliers:", total_outliers)
```

**HANDLING OUTLIERS**

```python
from scipy.stats.mstats import winsorize

# Define the proportion of outliers to keep (e.g., 0.05 means keep top 5% and bottom 5%)
proportion_to_keep = 0.05

# Apply winsorization to columns with outliers
columns_with_outliers = outliers_count[outliers_count > 0].index

for column in columns_with_outliers:
    lower_limit = proportion_to_keep / 2
    upper_limit = 1 - lower_limit
    data[column] = winsorize(data[column], limits=[lower_limit, upper_limit])


```

```python
# Calculate outlier counts in the winsorized dataset
winsorized_outliers_count = data[columns_with_outliers].apply(count_outliers)

# Compare original and winsorized outlier counts
outliers_comparison = pd.DataFrame({
    'Original Outliers': outliers_count[columns_with_outliers],
    'Winsorized Outliers': winsorized_outliers_count
})

print(outliers_comparison)
```

**FINE TUNING**

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Initialize the Gradient Boosting Classifier
gb_classifier = GradientBoostingClassifier(random_state=42)

# Define hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of boosting stages to be run
    'learning_rate': [0.01, 0.1, 0.2],  # Step size at each iteration
    'max_depth': [3, 4, 5],  # Maximum depth of the individual trees
}

# Initialize GridSearchCV
grid_search = GridSearchCV(gb_classifier, param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Fit the model on the training data with grid search
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_

# Initialize the model with the best hyperparameters
best_gb_classifier = GradientBoostingClassifier(**best_params, random_state=42)

# Fit the model on the training data
best_gb_classifier.fit(X_train, y_train)

# Predict on the test data
y_pred = best_gb_classifier.predict(X_test)

# Calculate accuracy
gr_accuracy = accuracy_score(y_test, y_pred)
print("Best Accuracy:", gr_accuracy)
```

**EVALUATION METRICS**

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report

# Create a confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Calculate precision, recall, and F1-score based on adjusted accuracy
precision = cm[1, 1] / (cm[1, 1] + cm[0, 1])
recall = cm[1, 1] / (cm[1, 1] + cm[1, 0])
f1_score = 2 * (precision * recall) / (precision + recall)

# Create a table of metrics
metrics_table = {
    'Metric': ['Precision', 'Recall', 'F1-Score', 'Accuracy'],
    'Value': [precision, recall, f1_score, gr_accuracy]
}

# Display the metrics table
metrics_df = pd.DataFrame(metrics_table)
print(metrics_df)


```
