**20.** Draw a plot between dep_delay and arr_delay

````python
import seaborn as sns

sns.set_style("darkgrid")
sns.set_context("paper")

sns.jointplot(x="dep_delay", y="arr_delay", data=flights, kind="hex", color="#4CB391")

sns.kdeplot(
    data=flights,
    x="dep_delay", y="arr_delay", 
    fill=True, cmap="YlGnBu", 
    levels=5, alpha=0.5, 
    thresh=0.1, linewidths=1.5,
)

sns.despine(left=True, bottom=True)


import matplotlib.pyplot as plt

plt.scatter(flights['dep_delay'], flights['arr_delay'])
plt.xlabel('Departure Delay')
plt.ylabel('Arrival Delay')
plt.title('JOINTPLOT  Departure Delay vs Arrival Delay')
plt.show()
````

It shows us that most of the data points are clustered around the bottom left corner of the plot, indicating that there is a strong correlation between the two variables. 

We can also see that the density of the data points decreases as we move towards the top right corner of the plot, indicating that there are fewer flights with higher departure and arrival delays. Finally, the hexagonal bins provide an easy way to see where most of the data points are located and how they are distributed.

````python
import matplotlib.pyplot as plt

plt.scatter(flights['dep_delay'], flights['arr_delay'])
plt.xlabel('Departure Delay')
plt.ylabel('Arrival Delay')
plt.title('SCATTER PLOT \n Departure Delay vs Arrival Delay')
plt.show()
````

````python
sns.jointplot(x='dep_delay', y='arr_delay', data=flights, kind='hex', cmap='YlOrRd')

# Add titles and axis labels
plt.suptitle('HEXBIN PLOT  Relationship Between Departure Delay and Arrival Delay', fontsize=16)
plt.subplots_adjust(top=0.95)
plt.xlabel('Departure Delay (minutes)', fontsize=12)
plt.ylabel('Arrival Delay (minutes)', fontsize=12)

# Show the plot
plt.show()
````

The resulting plot is a hexbin plot that shows the density of points for different combinations of departure and arrival delays. The color scale shows the density of points, with darker colors indicating higher density. 

The plot allows us to see that there is a positive correlation between departure delay and arrival delay, with many flights experiencing delays of up to 2 hours in both directions.
